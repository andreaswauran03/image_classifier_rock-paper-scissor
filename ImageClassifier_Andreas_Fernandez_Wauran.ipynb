```python
import os
from keras.preprocessing.image import ImageDataGenerator
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import foto
dataset_dir = "rockpaperscisor_dataset/"
```

```python
os.listdir(dataset_dir)
```

```python
# Inisiasi data augmentasi dengan ImageDataGenerator
datagen = ImageDataGenerator(rescale=1./255,
                             rotation_range=90,
                             shear_range=0.5,
                             zoom_range=0.1,
                             horizontal_flip=True,
                             fill_mode="nearest",
                             validation_split=0.4)
```

```python
# Konstanta pixel
SIZE = 200
```

Data Training

```python
# Membuat generator untuk data training
train_generator = datagen.flow_from_directory(
    directory=dataset_dir,
    target_size=(SIZE, SIZE),
    batch_size=25,
    class_mode='categorical',
    subset='training'
)
```

```python
# Inisialisasi list untuk menyimpan data dan label
datas = []
labels = []

# Mengiterasi melalui generator dan mengumpulkan data dan label
for i in range(len(train_generator)):
    batch_data, batch_labels = train_generator[i]
    datas.append(batch_data)
    labels.append(batch_labels)

# Menggabungkan batch-batch menjadi satu array besar
X_train = np.vstack(datas)
y_train_ohe = np.vstack(labels)
y_train = np.argmax(y_train_ohe, axis=1)
```

Data Validasi

```python
# Membuat generator untuk data validasi
validation_generator = datagen.flow_from_directory(
    directory=dataset_dir,
    target_size=(SIZE, SIZE),
    batch_size=10,
    class_mode='categorical',
    subset='validation'
)
```

```python
# Inisialisasi list untuk menyimpan data dan label
datas = []
labels = []

# Mengiterasi melalui generator dan mengumpulkan data dan label
for i in range(len(validation_generator)):
    batch_data, batch_labels = validation_generator[i]
    datas.append(batch_data)
    labels.append(batch_labels)

# Menggabungkan batch-batch menjadi satu array besar
X_validation = np.vstack(datas)
y_validation_ohe = np.vstack(labels)
y_validation = np.argmax(y_validation_ohe, axis=1)
```

Membangun Model

```python
from tensorflow.keras import models, layers, Sequential
```

```python
# Membuat model menggunakan model sequential
cnn = Sequential([
    # CNN
    layers.Conv2D(32, (3, 3), activation="relu", input_shape=(SIZE, SIZE, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation="relu"),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(128, (3, 3), activation="relu"),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(256, (3, 3), activation="relu"),
    layers.MaxPooling2D((2, 2)),

    # Dense
    layers.Flatten(),
    layers.Dense(256, activation="relu"),
    layers.Dense(3, activation="softmax")
])
```

```python
# Pengecekan Model yang dibangun
cnn.summary()
```

```python
# Compile model
cnn.compile(
    loss="sparse_categorical_crossentropy",
    optimizer="adam",
    metrics=["accuracy"])
```

```python
# Melatih model
cnn.fit(
    x=X_train,
    y=y_train,
    epochs=10,
    steps_per_epoch=200
)
```

```python
# Evaluasi Model
cnn.evaluate(X_validation, y_validation)
```

```python
# Prediksi Model
y_pred = cnn.predict(X_validation)
```

```python
# Konversi prediksi model menjadi indexing class nya
y_class = [np.argmax(element) for element in y_pred]
```

```python
classes = ["paper", "rock", "scissor"]
```

```python
# Melihat report dari prediksi model
from sklearn.metrics import classification_report
print(f"Classification Report: /n {classification_report(y_class, y_validation)}")
```

```python
# Membuat Plotting gambar untuk melihat prediksinya
def plot_img(x, y, index):
    plt.figure(figsize=(6,4))
    plt.imshow(x[index])
    plt.title(classes[y[index]])
    plt.show()
```

```python
plot_img(X_validation, y_validation, 0)
```
